### **1- Create a function: raw counts boxplot**
As we previously saw, raw counts can be accessed through ``counts(dds)`` line, it would be memory consuming, along with metadata information loss, to create an independant R object.

```r
library("RColorBrewer")
display.brewer.all() # to see palette's names
```

```r
library("DESeq2", quietly = TRUE)
```

```r
mycustomboxplot <- function(dataset = counts(dds.filteredbyN, normalized = FALSE),
                            samplesnum = 16, palette = "Greys",
                            transfolab = "Raw Counts", maintitle = "Filtered Raw Data"){
    
    colors <- colorRampPalette( rev(brewer.pal(9, palette)) )(samplesnum)

    boxplot(dataset, outline=F, col=colors, cex=1,
            xlab="samples", ylab=transfolab, xaxt="n", main=maintitle)
    axis(1, labels=seq(1,samplesnum,1), at=seq(1,samplesnum,1), xlab="Samples")
    
    return(invisible(NULL))
}
```

```r
mycustomboxplot(dataset = counts(dds, normalized = FALSE))
```

### **2- Descriptive transformation with base 2 logarithm (log2)**

> *To show the effect of the transformation, [...] we plot the first sample against the second, first simply using the log2 function (after adding 1, to avoid taking the log of zero), and then using the VST and rlog-transformed values. For the log2 approach, we need to first estimate size factors to account for sequencing depth, and then specify normalized=TRUE. Sequencing depth correction is done automatically for the vst and rlog.*  
[Love MI *et al.*, 2019](https://www.bioconductor.org/packages/release/workflows/vignettes/rnaseqGene/inst/doc/rnaseqGene.html#the-variance-stabilizing-transformation-and-the-rlog)

As the sequencing depth correction is automatically performed for the two others transformations, we will ask for it in following commandline:

```r
log2counts.filteredbyNnorm <- log2(counts(dds.filteredbyNnorm, normalized=TRUE) + 1)
```

```r
mycustomboxplot(dataset = log2counts.filteredbyNnorm,
                palette = "Reds",
                transfolab = "Log2(Normalized Counts +1)", maintitle = "Log2 Filtered Raw Data")
```

### **3- Variance stabilizing transformation (VST)**

> *In the above function calls, we specified `blind = FALSE`, which means that differences between cell lines and treatment (the variables in the design) will not contribute to the expected variance-mean trend of the experiment. The experimental design is not used directly in the transformation, only in estimating the global amount of variability in the counts. For a fully **unsupervised** transformation, one can set `blind = TRUE` (which is the default).*   
[Love MI *et al.*, 2019](https://www.bioconductor.org/packages/release/workflows/vignettes/rnaseqGene/inst/doc/rnaseqGene.html#the-variance-stabilizing-transformation-and-the-rlog)

```r
vsd.filteredbyNnorm <- vst(dds.filteredbyNnorm, blind = FALSE)
```

```r
mycustomboxplot(dataset = assay(vsd.filteredbyNnorm),
                palette = "Greens",
                transfolab = "Vst(Normalized Counts)",
                maintitle = "Variance Stabilizing Transformation (VST) on Filtered Normalized Data")
```

### **4- Regularized-logarithm transformation (rlog)**

```r
rlog.filteredbyNnorm <- rlog(dds.filteredbyNnorm, blind = FALSE)
```

> *I note that it takes longer time to rlog than for vst, but it is also limited to few minutes*

```r
mycustomboxplot(dataset = assay(rlog.filteredbyNnorm),
                palette = "Blues",
                transfolab = "Rlog(Normalized Counts)",
                maintitle = "Regularized-logarithm transformation (rlog) on Filtered Normalized Data")
```

### **5- Choosing best transformation for our data**

> *For genes with high counts, both the VST and the rlog will give similar result to the ordinary log2 transformation of normalized counts. For genes with lower counts, however, the values are shrunken towards a middle value. The VST or rlog-transformed data then become approximately homoskedastic (more flat trend in the meanSdPlot), and can be used directly for computing distances between samples, making PCA plots, or as input to downstream methods which perform best with homoskedastic data.  
> **Which transformation to choose?** The VST is much faster to compute and is less sensitive to high count outliers than the rlog. The rlog tends to work well on small datasets (n < 30), potentially outperforming the VST when there is a wide range of sequencing depth across samples (an order of magnitude difference). We therefore recommend the VST for medium-to-large datasets (n > 30). You can perform both transformations and compare the meanSdPlot or PCA plots generated, as described below.*  
[Love MI *et al.*, 2019](https://www.bioconductor.org/packages/release/workflows/vignettes/rnaseqGene/inst/doc/rnaseqGene.html#the-variance-stabilizing-transformation-and-the-rlog)

So, depending on our data set reduced to 16 samples, we could choose *rlog* transformation as there is also a *wide range of sequencing depth across samples* thanks to last sample sequenced twice more than the others.  

Both three tranformation methods performed well on my dataset, I notice that log2 boplot are less equal between samples than for the other two transformations. Medians don't seem to change between vst and rlog transformation.  

The **rlog** transformation keeps values for all pseudocounts range.  
Thus, I will keep this one to perform next steps: Principal Component Analysis (PCA) and Clustering (HeatMap).

## <b>II- Principal Component Analysis (PCA)</b>

### **1- Selecting most variant genes**

```r
# select the coefficient of variance from the selected transformed counts table
cv.calcul <- function(x){sd(x)/mean(x)}
```

Wanted number of most variant genes:

```r
ntop = 1000
```

Get a vector with coefficient of variance...

```r
cv <- apply(X = assay(rlog.filteredbyNnorm), MARGIN = 1, FUN = cv.calcul)
```

... we can order and use to select only the first highly variant features lines:

```r
select <- order(cv, decreasing = TRUE)[seq_len(min(ntop, length(cv)))]
mat.pca <- rlog.filteredbyNnorm[select, ]
```
